```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split

m = 150

def sigmoid(var):
    return 1 / (1 + np.exp(-var))

def ReLU(var):
    return np.maximum(var, 0)
    
def DeReLU(var):
    var[var < 0] = 0
    var[var >= 0] = 1
    return var

df = pd.read_csv('./iris.data', header=None)
X = df.iloc[:m, [0, 1, 2, 3]].values
Y = df.iloc[:m, [4]].values

for iter in range(m):
    if Y[iter][0] == 'Iris-setosa':
        Y[iter][0] = 1
    else:
        Y[iter][0] = 0

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.3, random_state = 0)

X_train = X_train.T
X_test = X_test.T
Y_train = Y_train.T
Y_test = Y_test.T
```

```python
'''
n0 = 4
n1 = 5
n2 = 4
n3 = 3
n4 = 2
n5 = 1
L = 5

x1  n10  n20
x2  n11  n21  n30  n40
    n12       n31       y_hat
x3  n13  n22  n32  n41
x4  n14  n23

'''

'''
L = 5
n = [4, 5, 4, 3, 2, 1]
'''

L = 2
n = [4, 5, 1]

W = [0, 0, 0, 0, 0, 0]
b = [0, 0, 0, 0, 0, 0]
dW = [0, 0, 0, 0, 0, 0]
db = [0, 0, 0, 0, 0, 0]
Z = [0, 0, 0, 0, 0, 0]
A = [0, 0, 0, 0, 0, 0]
dZ = [0, 0, 0, 0, 0, 0]
dA = [0, 0, 0, 0, 0, 0]

A[0] = X_train

for l in range(1, (L + 1)):
    W[l] = np.random.randn(n[l], n[(l - 1)]) * 0.01
    b[l] = np.zeros(shape = (n[l], 1))
    dW[l] = np.zeros(shape = (n[l], n[(l - 1)]))
    db[l] = np.zeros(shape = (n[l], 1))
    Z[l] = np.zeros(shape = (n[l], m))
    A[l] = np.zeros(shape = (n[l], m))
    dZ[l] = np.zeros(shape = (n[l], m))
    dA[l] = np.zeros(shape = (n[l], m))

for loop in range(10000):
    for l in range(1, (L + 1)):
        Z[l] = np.dot(W[l], A[l - 1]) + b[l]
        Z[l] = np.array(Z[l], dtype = np.float32)
        if l == L:
            A[l] = sigmoid(Z[l])
        else:
            A[l] = ReLU(Z[l])

    for l in range(L, 0, -1):
        if l == L:
            dZ[l] = A[l] - Y_train
        else:
            dZ[l] = dA[l] * (DeReLU(Z[l]))
        dW[l] = (np.dot(dZ[l], A[l - 1].T)) / m
        db[l] = (np.sum(dZ[l], axis = 1, keepdims = True)) / m
        dA[l - 1] = np.dot(W[l].T, dZ[l])
        W[l] = W[l] - dW[l]
        b[l] = b[l] - db[l]

A[0] = X_test
for l in range(1, (L + 1)):
    Z[l] = np.dot(W[l], A[l - 1]) + b[l]
    Z[l] = np.array(Z[l], dtype = np.float32)
    if l == L:
        A[l] = sigmoid(Z[l])
    else:
        A[l] = ReLU(Z[l])
        
print(Y_test - A[L])
```
